## EXPLORATORY DATA ANALYSIS

### Analysis By Country

```python
quantity_per_country = data.groupby(['Country'])['Quantity'].sum().reset_index()

#Arrange Country in Descending order
quantity_per_country = quantity_per_country.sort_values('Quantity', ascending = False).reset_index()
quantity_per_country.head(10)
```

```python
#Visualize Top 10 countries with highest number of quantity purchase
Top_10_countries = quantity_per_country.head(10)
plt.figure (figsize =(10,8))
ax = sns.barplot (x= "Country", y= "Quantity", data = Top_10_countries)
plt.title("Bar Chart of top 10 countries with highest purchase")
plt.xlabel('Countries')
plt.ylabel('Quantities of Product Sold')
plt.show()
```

```python
#Visualize countries with the least number of quantity purchase
Least_10_countries = quantity_per_country.tail(10)
plt.figure (figsize =(10,8))
ax = sns.barplot (x= "Country", y= "Quantity", data = Least_10_countries)
plt.title("Bar Chart of countries with the least purchase")
plt.xlabel('Country')
plt.ylabel('Quantity')
plt.show()
```

### Countries And No. of Customers Analysis

```python
customer_per_country = data.groupby(['Country'])['CustomerID'].nunique().reset_index()
customer_per_country = customer_per_country.sort_values('CustomerID', ascending = False).reset_index()

#rename column
customer_per_country.rename(columns = {"CustomerID" : "Number of customers"}, inplace = True)

customer_per_country.head(10)
```

```python
#visualize top 10 countries with highest number of customers (unique customerID)
top_countries_to_customers = customer_per_country.head(10)

# Create a bar chart from the selected data
plt.figure(figsize=(10,8))

sns.barplot(x='Country', y='Number of customers', data= top_countries_to_customers)

plt.xlabel('Countries')
plt.ylabel('Total number of customers')
plt.title('Bar Chart of top 10 countries with highest number of customers')
plt.show()
```
