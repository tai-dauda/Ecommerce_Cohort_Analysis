### Calculate Recency, Frequency, Monetary, and Customer Tenure

```python

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert InvoiceDate and CohortDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['CohortDate'] = pd.to_datetime(df['CohortDate'])

# Calculate Recency, Frequency, Monetary, and Customer Tenure

# Recency: Days since last purchase
last_purchase_date = df.groupby('CustomerID')['InvoiceDate'].max()
recency = (last_purchase_date.max() - last_purchase_date).dt.days

# Frequency: Count of transactions per customer
frequency = df.groupby('CustomerID')['InvoiceNo'].nunique()

# Monetary: Total spending per customer
df['TotalSpend'] = df['Quantity'] * df['UnitPrice']
monetary = df.groupby('CustomerID')['TotalSpend'].sum()

# Customer Tenure: Days from first to last purchase
first_purchase_date = df.groupby('CustomerID')['InvoiceDate'].min()
tenure = (last_purchase_date - first_purchase_date).dt.days

# Combine into a single DataFrame
rfm_tenure = pd.DataFrame({
    'Recency': recency,
    'Frequency': frequency,
    'Monetary': monetary,
    'Tenure': tenure
})

rfm_tenure.reset_index(inplace=True)
rfm_tenure

```

### Cohort Segmentation

```python

# Grouping by CohortDate and CohortIndex
cohort_group = data.groupby(['CohortDate', 'CohortIndex'])['CustomerID'].nunique().reset_index()

# Rename for clarity
cohort_group.columns = ['CohortDate', 'CohortIndex', 'NumCustomers']

# Display result
cohort_group.head(10)
```

```python
# Create a pivot table
cohort_pivot = cohort_group.pivot(index='CohortDate',
                                   columns='CohortIndex',
                                   values='NumCustomers')

# Reformat CohortDate index to a readable string format: "Month Year"
cohort_pivot.index = cohort_pivot.index.to_period('M').strftime('%b %Y')

# Display the pivot table with new index format
cohort_pivot.head(20)
```

```python
# Set up the figure size and style
plt.figure(figsize=(14, 8))
plt.title('Cohort Analysis - Retention of Customers', fontsize=16)

# Use seaborn's heatmap
sns.heatmap(cohort_pivot, 
            annot=True, 
            fmt='g',  # integer format
            cmap='YlGnBu', 
            linewidths=0.3,
            linecolor='gray')

plt.xlabel('Cohort Index (Months Since First Purchase)', fontsize=12)
plt.ylabel('Cohort Start Month', fontsize=12)
plt.yticks(rotation=0)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Get the size of each cohort (first column: Index = 1)
cohort_size = cohort_pivot.iloc[:, 0]

# Step 2: Divide each row by the cohort size to get retention rate
retention = cohort_pivot.divide(cohort_size, axis=0)

# Step 3: Plot the retention heatmap
plt.figure(figsize=(14, 8))
plt.title('Cohort Analysis - Retention Rate (%)', fontsize=16)

sns.heatmap(retention, 
            annot=True, 
            fmt='.0%',  # Show as percentage
            cmap='YlGnBu', 
            linewidths=0.3, 
            linecolor='gray',
            cbar=True)

plt.xlabel('Cohort Index (Months Since First Purchase)', fontsize=12)
plt.ylabel('Cohort Month (First Purchase)', fontsize=12)
plt.yticks(rotation=0)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Retention Matrix

```python
# Group data to calculate the average quantity per cohort and period
cohort_avg_quantity = data.groupby(['CohortDate', 'CohortIndex'])['Quantity'].mean().reset_index()

# Pivot the data into a cohort matrix with CohortDate as rows and CohortIndex as columns
cohort_pivot = cohort_avg_quantity.pivot(index='CohortDate', columns='CohortIndex', values='Quantity')

# Normalize the cohort data to get retention (first column is treated as base cohort size)
cohort_retention = cohort_pivot.divide(cohort_pivot.iloc[:, 0], axis=0) * 100

# Display the cohort retention matrix
cohort_retention
```

```python


# Prepare data for cohort analysis
df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')
df['CohortMonth'] = df.groupby('CustomerID')['CohortDate'].transform('min').dt.to_period('M')

# Calculate the cohort index (number of months since first purchase)
df['CohortIndex'] = ((df['InvoiceMonth'].dt.year - df['CohortMonth'].dt.year) * 12 + 
                     (df['InvoiceMonth'].dt.month - df['CohortMonth'].dt.month))

# Count unique customers by cohort and cohort index
cohort_data = df.groupby(['CohortMonth', 'CohortIndex'])['CustomerID'].nunique().reset_index()

# Pivot table for cohort analysis
cohort_pivot = cohort_data.pivot_table(index='CohortMonth', columns='CohortIndex', values='CustomerID')

# Calculate retention rate
cohort_sizes = cohort_pivot.iloc[:, 0]  # Number of customers in the first cohort month
retention_matrix = cohort_pivot.divide(cohort_sizes, axis=0) * 100  # Retention rate (%)

# Visualization of the retention matrix heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(retention_matrix, annot=True, fmt=".1f", cmap="Blues", cbar=True)
plt.title('Cohort Retention Rates')
plt.xlabel('Cohort Index (Months Since First Purchase)')
plt.ylabel('Cohort Month')
plt.show()

# Plotting retention curve for the first cohort
#first_cohort = retention_matrix.iloc[0]  # First cohort's retention curve
#first_cohort.plot(kind='line', marker='o', color='b', linewidth=2)
#plt.title('Retention Curve for First Cohort')
#plt.xlabel('Months Since First Purchase')
#plt.ylabel('Retention Rate (%)')
#plt.grid(True)
#plt.show()

plt.figure(figsize=(12, 8))
for cohort in retention_matrix.index:
    retention_matrix.loc[cohort].plot(label=str(cohort), marker='o', linewidth=2)

plt.title('Retention Curves for All Cohorts')
plt.xlabel('Months Since First Purchase')
plt.ylabel('Retention Rate (%)')
plt.legend(title='Cohort Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
